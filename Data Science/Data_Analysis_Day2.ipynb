```python
#Day1:(Data Analysis using pandas)
import pandas as pd
#Read csv file(csv files must be where notebook is)
df = pd.read_csv("avocado.csv")
df
albany_df=df[df["region"]=="Albany"]
albany_df.head()
#In older graph,date was running over each other,because pandas doesnt realise that its a date.
#Hence ,we need to explicitly assign date when we are reading csv file.
albany_df['Date']=pd.to_datetime(albany_df["Date"])
#index is basically the primary key by whuich all the rows can differentiate each other
df.index
#suppose i eant to change it to Date column
albany_df=albany_df.set_index("Date")  #Or albany_df.set_index("Date",inplace=True)

albany_df.head()
albany_df["AveragePrice"].plot()
#Suppose you want to label the xaxis and yaxis ,so you have to use pylot class of matplotlib and you ahve to import it .
import matplotlib.pyplot as plt
plt.ylabel("Average Price")
```

```python
#So,you can see its flunctuating very much.So,we can do is take rolling mean that is it will take mean of first 25 or n numbers and then plot it rather than plotting it individually.
albany_df["AveragePrice"].rolling(10).mean().plot()
```

```python
#its still flunctuating because data is not continous,tehrefore we can first sort index in order and them plot it.
albany_df.sort_index(inplace=True)
albany_df.index
albany_df["AveragePrice"].rolling(25).mean().plot() #it takes mean of 25 values .
```
